```python
weather['일시'] = weather['일시'].replace('.', '-')
weather.info()
```

```python
weather.rename(columns = {'일시': 'Date'}, inplace = True)
```

```python
weather
```

```python
weather.isnull().sum()
```

```python
weather.columns
```

```python
weather = weather[['Date', '평균기온(°C)', '최저기온(°C)', '최고기온(°C)', '최대 순간 풍속(m/s)', '최대 풍속(m/s)',
       '평균 풍속(m/s)', '최소 상대습도(%)', '평균 상대습도(%)']]
```

```python
weather['최저기온(°C)'] = weather['최저기온(°C)'].fillna(weather['최저기온(°C)'].mean())
```

```python
weather['최대 순간 풍속(m/s)'] = weather['최대 순간 풍속(m/s)'].fillna(weather['최대 순간 풍속(m/s)'].mean())
weather['최대 풍속(m/s)'] = weather['최대 풍속(m/s)'].fillna(weather['최대 풍속(m/s)'].mean())
weather['평균 풍속(m/s)'] = weather['평균 풍속(m/s)'].fillna(weather['평균 풍속(m/s)'].mean())
```

```python
weather
```

```python
weather['Date'] = pd.to_datetime(weather['Date'], format='%Y.%m.%d')

# 원하는 형식으로 변환
weather['Date'] = weather['Date'].dt.strftime('%Y-%m-%d')
```
